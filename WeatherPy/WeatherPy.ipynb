```python
from citipy import citipy
```

```python
import random


nearest_city_list =[]
for i in range(0,2000):
    random_lat = random.uniform(-90,90) 
    random_lng = random.uniform(-180,180)

    nearest_city = citipy.nearest_city(random_lat,random_lng)
    nearest_city_list.append(nearest_city.city_name)

```

## Generate Cities List

```python
unique_city_list = []
for city in nearest_city_list:
    if city not in unique_city_list:
        unique_city_list.append(city)
```

```python
len(unique_city_list)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
from api_keys import weather_api_key,g_key
import requests
import json
```

```python
url_base = "http://api.openweathermap.org/data/2.5/weather?"
print("Beginning Data Retrieval")
print("-----------------------------------")
city_real = []
lat = []
lng = []
max_temp = []
humidity = []
cloudness = []
wind_speed = []
country = []
date = []

for i in range(len(unique_city_list)):
    params = {"q":unique_city_list[i],
              "appid":weather_api_key,
             }

    response = requests.get(url_base,params = params).json()
    
    try:
        city_real.append(response["name"])
        city_number = city_real.index(response["name"]) + 1
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        max_temp.append(response["main"]["temp_max"])
        humidity.append(response["main"]["humidity"])
        cloudness.append(response["clouds"]["all"])
        wind_speed.append(response["wind"]["speed"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        
        print(f"Processing Record {city_number} of Set 1 | {response['name']}")

    except:
        print("City not found. Skipping...")
        
print("-----------------------------------")         
print("Data Retrieval Complete")
print("-----------------------------------")       
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
import pandas as pd
```

```python
city_df = pd.DataFrame({"City":city_real,
                       "Lat":lat,
                       "Lng":lng,
                       "Max Temp":max_temp,
                       "Humidity":humidity,
                       "Cloudiness":cloudness,
                       "Wind Speed":wind_speed,
                       "Country":country,
                       "Date":date})
```

```python
city_df.to_csv("../Resources/cities.csv",index = False)
```

## Linear Regression

```python
NH_df = city_df.loc[city_df['Lat']>=0,:]
SH_df = city_df.loc[city_df['Lat']<0,:]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
from scipy.stats import linregress
x_axis_NH = NH_df['Lat']
y_axis_NH_maxtemp = NH_df['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis_NH, y_axis_NH_maxtemp)
regress_values = x_axis_NH * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_axis_NH,y_axis_NH_maxtemp,edgecolors='black')
plt.plot(x_axis_NH,regress_values,"r-")
plt.annotate(line_eq,(10,270),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temp (F)')
print(f"The r-value is: {rvalue}")
print("There is a strong negative relationship between latitude and max temperature for northern hemisphere.")
plt.savefig("Output/latitude and max temperature for northern hemisphere.png")
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python

x_axis_SH = SH_df['Lat']
y_axis_SH_maxtemp = SH_df['Max Temp']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis_SH, y_axis_SH_maxtemp)
regress_values = x_axis_SH * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_axis_SH,y_axis_SH_maxtemp,edgecolors='black')
plt.plot(x_axis_SH,regress_values,"r-")
plt.annotate(line_eq,(-50,305),fontsize=15,color="red")
plt.xlabel('Latitude')
plt.ylabel('Max Temp (F)')
print(f"The r-value is: {rvalue}")
print("There is a strong positive relationship between latitude and max temperature for southern hemisphere.")
plt.savefig("Output/latitude and max temperature for southern hemisphere.png")
```
